```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
import pandas as pd
from mpl_toolkits.basemap import Basemap
```

```python
#Set up map
plt.figure(figsize=(15,15))
m = Basemap(projection='merc',lon_0=-103,lat_0=35,resolution='l',
           llcrnrlon=-128, llcrnrlat=20,
           urcrnrlon=-60, urcrnrlat=53)
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
#Draw Country and State Lines
m.drawmapboundary(fill_color='aqua')
m.drawcountries(linewidth=1)
m.drawstates(linewidth=0.5)

#Draw Meridians and Parallels
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])
parallels = np.arange(0.,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])

#Getting lat / longs for impact sites
impacts = pd.read_csv('Meteorite_Landings.csv')
#print(impacts.count)

#Remove the any rows that had a column that was left empty
impacts = impacts.dropna(how='any')
#print(impacts.count)

#Put lat and longs into arrays
lats = impacts['reclat'].values
longs = impacts['reclong'].values

#Scatter - Markers for impact sites
m.scatter(longs, lats, marker='D',color='m', latlon=True) #Home
plt.title("Meteorite Impacts USA")
plt.show()
```

```python
#West Count of Impacts
sumWest = sum((impacts['reclong'] >= -125) & (impacts['reclong'] <= -110)
                  & (impacts['reclat'] >= 24) & (impacts['reclat'] <= 50))

print('West Count: ', sumWest)

#Mid-West Count of Impacts
sumMidWest = sum((impacts['reclong'] >= -109) & (impacts['reclong'] <= -91)
                  & (impacts['reclat'] >= 24) & (impacts['reclat'] <= 50))

print('Mid-West Count: ', sumMidWest)

#East Count of Impacts
sumEast = sum((impacts['reclong'] <= -65) & (impacts['reclong'] >= -90)
                  & (impacts['reclat'] >= 24) & (impacts['reclat'] <= 50))

print('East Coast Count: ', sumEast)
```
